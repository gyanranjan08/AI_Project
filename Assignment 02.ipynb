```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt

#load the dataset
digits=load_digits()

#Dataset imformation
print("Dataset Shape:",digits.data.shape)
print("Image shape:",digits.images.shape)
print("Target Labels:",digits.target_names)

#Display the first image
plt.figure(figsize=(4,4))
plt.imshow(digits.images[0],cmap='pink')
plt.title(f"label:{digits.target[0]}")
plt.axis('off')
plt.show()
```

****Q2. write a program to display a. number of rows and columns,  b. column names and c. data types****

```python
#import Libraries
import pandas as pd
from sklearn.datasets import load_iris

#load the iris dataset
iris=load_iris(as_frame=True)

#convert to pandss DataFrame
df=iris.frame


#Display the number of rows and columns

print(df.shape)


#Display the column names
print("\nColumn Names:")
print(df.columns)


#Display the data types
print("\nData Types:")
print(df.dtypes)
```

****Q3. write a program to count the number of samples in iris dataset and digit dataset.****

```python
#import datasets
from sklearn.datasets import load_iris, load_digits

# Load datasets
iris = load_iris()
digits = load_digits()

# Count samples
iris_samples = len(iris.data)
digits_samples = len(digits.data)

print("Number of samples in Iris dataset:", iris_samples)
print("Number of samples in Digits dataset:", digits_samples)
```
